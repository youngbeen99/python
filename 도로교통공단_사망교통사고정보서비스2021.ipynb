```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' with columns 'injpsn_cnt' and 'dth_dnv_cnt'

# Calculate the seriously injured count
df['seriously_injured_cnt'] = df['injpsn_cnt'] - df['dth_dnv_cnt']

# Calculate the total sum of each category
total_injured = df['injpsn_cnt'].sum()
total_deceased = df['dth_dnv_cnt'].sum()
total_seriously_injured = df['seriously_injured_cnt'].sum()

# Create a list of values for each category
values = [total_injured, total_deceased, total_seriously_injured]

# Create labels for each category
labels = ['부상자', '사망자', '중상자']

# Set colors for each category
colors = ['blue', 'red', 'green']

# Create a pie chart
plt.pie(values, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Set aspect ratio to make it a circle
plt.axis('equal')

# Add a title
plt.title('교통사고 통계')

# Show the chart
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' with columns 'dth_dnv_cnt', 'injpsn_cnt', 'seriously_injured_cnt'

# Calculate the total sum of each category
total_deceased = df['dth_dnv_cnt'].sum()
total_injured = df['injpsn_cnt'].sum()
total_seriously_injured = df['seriously_injured_cnt'].sum()

# Create a list of values for each category
values = [total_deceased, total_injured, total_seriously_injured]

# Create labels for each category
labels = ['사망자', '부상자', '중상자']

# Calculate angles for each category
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]

# Repeat the first value of `values` at the end to match the length of `angles`
values += values[:1]

# Create a radar chart
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, values, 'o-', linewidth=2)  # Connect the points with lines

# Fill the area under the radar chart
ax.fill(angles, values, alpha=0.25)

# Set the angle ticks and labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# Set the y-axis label
ax.set_ylabel('Count')

# Set the title
ax.set_title('교통사고 통계')

# Display the radar chart
plt.show()
```
